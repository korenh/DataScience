# Distributions

Statistics / Data distributions other than theoretical knowledge, helps to be able to use them in modeling, data analysis, insights etc.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Uniform Distribution
In Uniform Distribution, values within a specific range are equally likely to occur & Values outside that given range never occur

```python
sample = np.random.uniform(0, 1, 10000)
plt.hist(sample, bins=np.arange(-1,2,0.05)) 
plt.title('Uniform Distribution'), plt.axis('off')
```

### Normal/Gaussian Distribution
Normal/Gaussian Distribution is symmetric distribution about its mean where most of the observations cluster around the mean and the probabilities for values further away from the mean taper off equally in both directions

```python
sample = np.random.normal(0, 1, 10000)
plt.hist(sample, bins=np.arange(-5,5,0.1)) 
plt.title('Normal/Gaussian Distribution'), plt.axis('off')
```

### Binomial Distribution
Binomial Distribution expresses the probability of a given number of successes in a sequence of n independent experiments with a known probability of success on each trial

```python
sample = np.random.binomial(20, 0.5, 10000)
plt.hist(sample, bins=np.arange(0,20,1)) 
plt.title('Binomial Distribution'), plt.axis('off')
```

### Geometric Distribution
Geometric Distribution expresses the probability distribution of the random variable (X) representing number of Bernoulli trials needed to get one success

```python
sample = np.random.geometric(0.5, 1000)
plt.hist(sample, bins=np.arange(0,20,1)) 
plt.title('Geometric Distribution'), plt.axis('off')
```

### Exponential Distribution
Exponential distribution is the probability distribution of the time between events in a Poisson point process, a process in which events occur continuously and independently at a constant average rate. It is a particular case of the gamma distribution

```python
sample = np.random.exponential(1, 10000)
plt.hist(sample, bins=np.arange(0,10,0.1))
plt.title('Exponential Distribution'), plt.axis('off')
```

### Poisson Distribution
Poisson Distribution is a discrete probability distribution and it expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event

```python
sample = np.random.poisson(10, 10000)
plt.hist(sample, bins=np.arange(0,20,1)) 
plt.title('Poisson Distribution'), plt.axis('off')
```

### Logistic Distribution
Logistic distribution is a continuous probability distribution. Its cumulative distribution function is the logistic function, which appears in logistic regression and feedforward neural networks. It resembles the logistic distribution in shape but has heavier tails

```python
sample = np.random.logistic(0, 1, 10000)
plt.hist(sample, bins=np.arange(-5,5,0.1)) 
plt.title('Logistic Distribution'), plt.axis('off')
```
