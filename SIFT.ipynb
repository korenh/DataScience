# SIFT
Scale Invariant Feature Transform is a feature detection algorithm.
Main advantage of SIFT features is that they not affected by image size or orientation

### Imports

```python
import cv2 
import matplotlib.pyplot as plt
```

### PreProcess
Read image, resize & grayscale conversion 

```python
img1 = cv2.resize(cv2.imread('./Data/SIFT/sample1.png'), (600, 400))
img2 = cv2.resize(cv2.imread('./Data/SIFT/sample2.png') , (600, 400))

gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

### Features

```python
sift = cv2.xfeatures2d.SIFT_create()

keypoints_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2,None)

img_1 = cv2.drawKeypoints(gray1,keypoints_1,img1)
plt.figure(figsize=(10, 10)),plt.imshow(img1)
```

### Match

```python
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)
img3 = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:50], img2, flags=2)

plt.figure(figsize=(20, 20)),plt.imshow(img3)
```
