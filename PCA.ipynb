# PCA 
Principal Component Analysis is a mathematical technique used for dimensionality reduction.
#### Benefits
- Reduces training time - due to smaller dataset.
- Removes noise - by keeping only whats relevant.
- Makes visualization possible - in cases where you have a maximum of 3 principal components.

### Imports

```python
import numpy as np 
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

### Dataset
Iris dataset contain features - sepal_length, sepal_width, petal_length, petal_width	

```python
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

### Scale
Standardize features by removing the mean and scaling to unit variance.

```python
data_scaled = StandardScaler().fit_transform(df.drop(columns=['species']))
```

### Covariance matrix

Provide strength measurment of correlation.

```python
features = data_scaled.T
cov_matrix = np.cov(features)
```

### Eigendecomposition
Decompose matrix into eigenvectors and eigenvalues.

```python
values, vectors = np.linalg.eig(cov_matrix)
```

### Visualization

```python
v1 = data_scaled.dot(vectors.T[0])
v2 = data_scaled.dot(vectors.T[1])

sns.set(rc={'figure.figsize':(15, 10)})
sns.scatterplot(x = v1, y = v2, hue=df['species'], s=200)
```
