# SVM

SVM is supervised algorithm works best on smaller datasets but on complex ones. Support Vector Machine, abbreviated as SVM can be used for both regression and classification

### Imports

```python
from sklearn import svm
from sklearn import datasets
import matplotlib.pyplot as plt
```

### Dataset
Load digits dataset from skelearn 

```python
dataset = datasets.load_digits()
```

### Classifier


- gamma - Kernel coefficient for ‘rbf’, ‘poly’ and ‘sigmoid’

- C - Regularization parameter, inversely proportional to C

```python
classifier = svm.SVC(gamma=0.001, C=100)
```

### Preprocess

Loads in all but the last 10 data points, so we can them for testing. 

```python
data ,labels  = dataset.data[:-10], dataset.target[:-10]
```

### Train

```python
classifier.fit(data, labels)
```

### Test 

```python
idx = 500
plt.title(str(classifier.predict(dataset.data[idx].reshape(1,-1))))
plt.imshow(dataset.images[idx])
```
